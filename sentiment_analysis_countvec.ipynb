```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv('data.txt', sep = '\t')
```

```python
df
```

### **STEP 1: PREPROCESSING**

```python
# Download NLTK resources
nltk.download('stopwords')
nltk.download('punkt')

def preprocess_tweet(tweet):
    # Convert to lowercase
    tweet = tweet.lower()
    
    # Remove URLs
    tweet = re.sub(r"http\S+|www\S+|https\S+", "", tweet, flags=re.MULTILINE)
    
    # Remove special characters and digits
    tweet = re.sub(r"[^\w\s]", "", tweet)
    tweet = re.sub(r"\d+", "", tweet)
    
    # Tokenize the tweet
    tokens = word_tokenize(tweet)
    
    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    tokens = [word for word in tokens if word not in stop_words]
    
    # Join tokens back into a single string
    preprocessed_tweet = " ".join(tokens)
    
    return preprocessed_tweet

```

```python
df['pre_tweet'] = df['Tweet'].apply(preprocess_tweet)
```

```python
df
```

```python
df = df.drop(['ID','Tweet'], axis = 1)
```

```python
columns = df.columns.tolist()
columns = [columns[-1]] + columns[:-1:]
df = df[columns]
df
```

```python
sent_list = df.columns.tolist()
sent_list.remove('pre_tweet')
print(sent_list)
```

## **STEP 2: APPLYING COUNT VECTORIZER**

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Assuming your DataFrame is named 'df' and has a 'tweet' column and multiple sentiment columns
tweets = df['pre_tweet'].values
sentiments = df.drop('pre_tweet', axis=1).values

# Split the data into training and testing sets
train_tweets, test_tweets, train_sentiments, test_sentiments = train_test_split(tweets, sentiments, test_size=0.2, random_state=42)

# Preprocess the tweets (e.g., tokenization, stemming, or lemmatization)
# You can perform further preprocessing steps based on your requirements

# Vectorize the preprocessed tweets using a CountVectorizer
vectorizer = CountVectorizer()
train_features = vectorizer.fit_transform(train_tweets)
test_features = vectorizer.transform(test_tweets)


```

## **STEP 3: FITTING THE MULTINOMIAL NAIVE BAYES CLASSIFIER MODEL**

```python
# Train the Multinomial Naive Bayes classifier for each sentiment separately
classifiers = []
for i in range(train_sentiments.shape[1]):
    classifier = MultinomialNB()
    classifier.fit(train_features, train_sentiments[:, i])
    classifiers.append(classifier)

# Predict the sentiments of the test data for each classifier
predicted_sentiments = []
for classifier in classifiers:
    predicted_sentiments.append(classifier.predict(test_features))
    
# Convert the predicted sentiments to the original shape
predicted_sentiments = pd.DataFrame(predicted_sentiments).T.values
```

## **EVALUATING THE MODEL**

```python
# Evaluate the accuracy of the classifier for each sentiment separately
accuracies = []
for i in range(test_sentiments.shape[1]):
    accuracy = accuracy_score(test_sentiments[:, i], predicted_sentiments[:, i])
    accuracies.append(accuracy)

# Calculate the average accuracy across all sentiments
average_accuracy = sum(accuracies) / len(accuracies)
print("Average Accuracy:", average_accuracy)
```
